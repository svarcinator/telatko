```python
import spot
from spot.jupyter import display_inline
spot.setup(show_default='.bans')
```

```python
def show_one_difference(left, right):
    run = left.exclusive_run(right)
    if not run:
        print("The two automata are equivalent.")
    else:
        print("The following word is only accepted by one automaton:", spot.make_twa_word(run))
        run.highlight(5)
        display_inline(left, right)
```

```python
for aut in spot.automata('./acd1_optimized.hoa'):
    display(aut)
```

```python

for orig in spot.automata('./acd1.hoa'):
    display(orig)
```

```python
for scc in spot.automata('./acd1_alternative.hoa'):
    display(scc)
```

```python
for scc_res in spot.automata('./acd1_alternative_res.hoa'):
    display(scc_res)
```

```python
show_one_difference(scc, scc_res)
```
